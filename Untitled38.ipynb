```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/content/Student_Performance.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt

y = df.groupby('gender')['study_hours'].std()

plt.plot(y.index, y.values, marker='o')
plt.xlabel("Gender")
plt.ylabel("Std of Study Hours")
plt.title("Standard Deviation of Study Hours by Gender")
plt.show()
```
